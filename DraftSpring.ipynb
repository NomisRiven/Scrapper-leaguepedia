```python
TOURNAMENTS = ["LCK 2026 Rounds 1-2", "LPL 2026 Split 2", "LEC 2026 Spring"]
PATCHES     = None          # ex: ["25.4", "25.5"] or None for all
TEAM        = None        # ex: "SK Gaming" or None for all
LIMIT       = 500          # ex: 100 or None for all

# ── Auto filename builder ────────────────────────────────────────────────────
# Produces a suffix like: LEC2026Versus-LCKCup2026_SKGaming_p25.4-25.5
# Every section below uses it automatically.

def _build_suffix(tournaments=None, team=None, patches=None):
    parts = []
    if tournaments:
        parts.append("-".join(t.replace(" ", "") for t in tournaments))
    if team:
        parts.append(team.replace(" ", ""))
    if patches:
        parts.append("p" + "-".join(patches))
    return "_".join(parts)

SUFFIX = _build_suffix(TOURNAMENTS, TEAM, PATCHES)
print(f"📁 File suffix: {SUFFIX}")
```

```python
# (team, pick_slot, opponent_team, opponent_slots_already_revealed)
pick_order = [
    ('Team1', 1, 'Team2', []),          # T1P1: T2 showed nothing
    ('Team2', 1, 'Team1', [1]),         # T2P1: T1 showed pick 1
    ('Team2', 2, 'Team1', [1]),         # T2P2: T1 still only showed pick 1
    ('Team1', 2, 'Team2', [1, 2]),      # T1P2: T2 showed picks 1+2
    ('Team1', 3, 'Team2', [1, 2]),      # T1P3: T2 showed picks 1+2
    ('Team2', 3, 'Team1', [1, 2, 3]),   # T2P3: T1 showed picks 1+2+3
    ('Team2', 4, 'Team1', [1, 2, 3]),   # T2P4: ban phase, T1 still at 1+2+3
    ('Team1', 4, 'Team2', [1, 2, 3, 4]),
    ('Team1', 5, 'Team2', [1, 2, 3, 4]),
    ('Team2', 5, 'Team1', [1, 2, 3, 4, 5]),
]

rows = []
for team, slot, opp, opp_slots in pick_order:
    champ_col = f'{team}Pick{slot}'
    role_col  = f'{team}Role{slot}'
    opp_role_cols = [f'{opp}Role{s}' for s in opp_slots]

    for _, row in df.iterrows():
        champ = row[champ_col]
        role  = row[role_col]
        if champ == '':
            continue

        opp_roles_revealed = [row[c] for c in opp_role_cols]
        is_blind = role not in opp_roles_revealed

        rows.append({
            'Champion': champ,
            'Role':     role,
            'Team':     team,
            'Slot':     slot,
            'IsBlind':  is_blind
        })

result_df = pd.DataFrame(rows)

blind_by_team = result_df[result_df['IsBlind']].groupby('Team')['IsBlind'].count()
print(blind_by_team)

# Or as a rate per game
total_picks_by_team = result_df.groupby('Team')['IsBlind'].count()
blind_rate = result_df[result_df['IsBlind']].groupby('Team')['IsBlind'].count() / total_picks_by_team
print(blind_rate)
```
