## ⚙️ Config

```python
#TOURNAMENTS = ["LEC 2026 Versus", "LEC 2026 Versus Playoffs", "LEC 2026 Spring"]
#TOURNAMENTS = ["LEC 2026 Spring"]
TOURNAMENTS = ["LFL 2026 Invitational", "LFL 2026 Spring", "EM 2026 Winter", "Esports World Cup 2026 Online Qualifier: EMEA"]

PATCHES     = None          # ex: ["25.4", "25.5"] or None for all
TEAM        = "Galions"          # ex: "SK Gaming" or None for all
#TEAM = None
LIMIT       = 500          # ex: 100 or None for all

# ── Auto filename builder ────────────────────────────────────────────────────
# Produces a suffix like: LEC2026Versus-LCKCup2026_SKGaming_p25.4-25.5
# Every section below uses it automatically.

def _build_suffix(tournaments=None, team=None, patches=None):
    parts = []
    if tournaments:
        parts.append("-".join(t.replace(" ", "").replace(":", "_") for t in tournaments))
    if team:
        parts.append(team.replace(" ", ""))
    if patches:
        parts.append("p" + "-".join(patches))
    return "_".join(parts)

SUFFIX = _build_suffix(TOURNAMENTS, TEAM, PATCHES)
print(f"📁 File suffix: {SUFFIX}")
```

```python
df = scrape_draft_data(site, TOURNAMENTS, limit=500)
df_filtered = filter_dataframe(df, team_filter=TEAM)
df_filtered
```

```python
FirstPick_List = df_filtered[
    (df_filtered['Team1'] == TEAM) & 
    (df_filtered['Team1Pick1'] != '')
]['Team1Pick1']

counts = FirstPick_List.value_counts()
counts
```

```python
FirstPick_List_G1 = df_filtered[
    (df_filtered['Team1'] == TEAM) & 
    (df_filtered['Team1Pick1'] != '') &
    (df_filtered['GameId'].str.endswith('_1'))
]['Team1Pick1']

counts = FirstPick_List_G1.value_counts()
counts
```

## 1 · Draft cards
Visualise every draft as a card. Set `FILTER_CHAMPION` / `HIGHLIGHT_CHAMPION` as needed.

```python
FILTER_CHAMPION    = None   # ex: "Azir"  → only games where Azir appears
HIGHLIGHT_CHAMPION = None   # ex: "Azir"  → highlight Azir in yellow

_suffix = _build_suffix(TOURNAMENTS, TEAM, PATCHES)
if FILTER_CHAMPION:
    _suffix += f"_{FILTER_CHAMPION}"

analyze_drafts(
    site,
    tournaments        = TOURNAMENTS,
    output_filename    = f"drafts_{_suffix}.html",
    team_filter        = TEAM,
    patches            = PATCHES,
    filter_champion    = FILTER_CHAMPION,
    highlight_champion = HIGHLIGHT_CHAMPION,
    limit              = 500,
)
```
