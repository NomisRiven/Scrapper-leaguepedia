```python
import requests
import pandas as pd

# Version actuelle du jeu (tu peux la fixer manuellement si besoin)
version = requests.get("https://ddragon.leagueoflegends.com/api/versions.json").json()[0]

# URL des champions
url = f"https://ddragon.leagueoflegends.com/cdn/{version}/data/en_US/champion.json"

data = requests.get(url).json()

# Extraction des noms (typo officielle)
champions = [champ["name"] for champ in data["data"].values()]

# Optionnel : trier
champions = sorted(champions)

# Format Google Sheets (une colonne)
df = pd.DataFrame(champions, columns=["Champion"])

# Affichage
df
```
